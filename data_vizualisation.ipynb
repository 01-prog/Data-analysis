```python
import pandas as pd

import matplotlib.pyplot as plt

# Extract price and query from data
records = []
for item in data:
    price = float(item['prix']['amount'])
    query = item['query']
    records.append({'query': query, 'price': price})

df = pd.DataFrame(records)
# Map 'query' to combined category string using category_mapping
df['category'] = df['query'].map(lambda q: category_mapping[q][0] + ' ' + category_mapping[q][1] if q in category_mapping else 'Unknown')

# Calculate mean price per category
mean_prices = df.groupby('category')['price'].mean()

# Plot horizontal bar graph
mean_prices_sorted = mean_prices.sort_values()
mean_prices_sorted.plot(kind='barh', figsize=(10, 8))
plt.xlabel('Mean Price (EUR)')
plt.ylabel('Category')
plt.title('Mean Price per Category')
plt.show()
```
